````python
import numpy as np
import matplotlib.pyplot as plt

#  
fs = 30000  #  
f = 400    #  
duration = 1  #    
t = np.arange(0, duration, 1/fs)  #  

#   
A = 1  #  
x = A * np.sin(2*np.pi*f*t)

#   
sigma = 0.1  #   
noise = sigma * np.random.randn(len(x))
y = x + noise

#  
plt.figure(figsize=(10, 6))
plt.plot(t, y)
plt.title("    ")
plt.xlabel(", ")
plt.ylabel("")
plt.grid()
plt.show()
````
